```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
URL = 'https://www.unique-canvas.com/artprints/art_styles.html'

response = requests.get(URL)
soup = BeautifulSoup(response.text, 'html.parser')

link_container = soup.findAll('div', class_= 'sub-category-container row')[0]
```

```python
current_listing = link_container.findAll('a', class_='sub-link')

current_links = []

# get url page for each modern current
for current in current_listing:
    current_links.append(current['href'])

current_links
```

```python
current_list = []
current_description = []

# iterate through each link to get name of current and it's description
for link in current_links:
    response = requests.get(link)
    soup = BeautifulSoup(response.text, 'html.parser')

    current_title = soup.findAll('ul','li', class_='items')[0]
    title = current_title.findAll('li', class_='item')
    
    description = soup.find('div', class_= 'cat-desc')
    
    # remove h2 title from description part
    unwanted = description.find('h2')
    unwanted.extract()
    
    current_list.append(title[-1].getText())
    current_description.append(description.getText().strip())

    
    print('CURRENT TITLE :', title[-1].getText())
    print('DESCRIPTION :', description.getText())
    print('-----')
```

```python
# create and export dataframe
df = pd.DataFrame({"current" : current_list, "current_description": current_description})
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
df['id_current'] = le.fit_transform(df['current']) + 1
```
